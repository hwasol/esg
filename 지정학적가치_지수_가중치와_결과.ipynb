```python
import pandas as pd

# 파일명에 맞춰서 불러오기 (모두 '국가', '2022' 컬럼만)
df_un = pd.read_csv('아프리카_25개국_UN평화유지군_파견인원_2011-2022.csv')[['국가', '2022']]
df_trade = pd.read_csv('아프리카_25개국_국제무역규모_2011-2022.csv')[['국가', '2022']]
df_border = pd.read_csv('아프리카_국경공유국수_2011-2022.csv')[['국가', '2022']]
df_regional = pd.read_csv('아프리카_지역기구활동점수_2011-2022.csv')[['국가', '2022']]
df_terror = pd.read_csv('아프리카_테러발생건수_2011-2022.csv')[['국가', '2022']]

# 각 파일별 컬럼명 통일
df_un.columns = ['국가', 'UN파견']
df_trade.columns = ['국가', '국제무역']
df_border.columns = ['국가', '국경공유']
df_regional.columns = ['국가', '지역기구']
df_terror.columns = ['국가', '테러']

# 국가명을 기준으로 머지 (순서는 상관없음)
df = df_un.merge(df_trade, on='국가') \
          .merge(df_border, on='국가') \
          .merge(df_regional, on='국가') \
          .merge(df_terror, on='국가')

display(df)  # 데이터 병합 결과 미리보기
```

```python
from sklearn.preprocessing import StandardScaler

features = ['UN파견', '국제무역', '국경공유', '지역기구', '테러']

# 결측치가 있을 경우(혹시 모르니) 평균값으로 채우기
df[features] = df[features].fillna(df[features].mean())

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])
```

```python
from sklearn.decomposition import PCA
import numpy as np

# PCA 실행
pca = PCA()
pca.fit(X_scaled)

# 첫 번째 주성분의 로딩값(가중치) 추출
weights = pca.components_[0]

# 각 변수별 가중치 출력
for feat, w in zip(features, weights):
    print(f'{feat}: {w:.4f}')

# 첫 번째 주성분이 전체 분산의 몇 %를 설명하는지 확인
print('\n첫 번째 주성분 설명력:', f"{pca.explained_variance_ratio_[0]:.2%}")
```

```python
# 국가별 지정학적 가치 지수(첫 번째 주성분 점수)
df['지정학적가치'] = np.dot(X_scaled, weights)

# 결과 미리보기
display(df[['국가', '지정학적가치']])

# (선택) 결과 CSV로 저장
df[['국가', '지정학적가치']].to_csv('아프리카_지정학적가치_PCA결과.csv', index=False)

# (선택) 변수별 가중치 저장
pd.DataFrame({'지표명': features, 'PCA_가중치': weights}).to_csv('PCA_가중치표.csv', index=False)
```

```python
# 지정학적 가치 결과 저장
df[['국가', '지정학적가치']].to_csv('아프리카_지정학적가치_PCA결과.csv', index=False, encoding='utf-8-sig')

# 가중치표 저장
pd.DataFrame({'지표명': features, 'PCA_가중치': weights}).to_csv('PCA_가중치표.csv', index=False, encoding='utf-8-sig')
```
